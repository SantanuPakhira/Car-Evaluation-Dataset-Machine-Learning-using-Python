```python
data_dummies=pd.get_dummies(data=car,columns= ['2','2.1','small','low','vhigh','vhigh.1'
], \
                           prefix = ['2','2.1','small','low','vhigh','vhigh.1'])

data_dummies.head()
```

```python
from sklearn.model_selection import train_test_split
# Train-Test split: 20% test data
Cars = data_dummies.drop('unacc', 1)
label = data_dummies.unacc
data_train, data_test, label_train, label_test = train_test_split(Cars, label, test_size = 0.2, random_state = 50)
```

```python
from sklearn import tree
# Decision tree with depth = 2
dt2 = tree.DecisionTreeClassifier()
dt2.fit(data_train, label_train)
dt2_score_train = dt2.score(data_train, label_train)
print("Training score: ",dt2_score_train)
dt2_score_test = dt2.score(data_test, label_test)
print("Testing score: ",dt2_score_test)
```
